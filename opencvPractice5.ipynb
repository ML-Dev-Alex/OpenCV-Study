```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('nighthawks.jpg')

cv.imshow("original", img)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(img, dsize=(450, 200), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])

```

```python
plt.clf()
plt.cla()
plt.close()

laplacian = cv.Laplacian(img, cv.CV_16S, ksize=1)

cv.imshow("laplacian", laplacian)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(laplacian, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

sobelx = cv.Sobel(img, cv.CV_16S, 1, 0, ksize=3)

cv.imshow("sobelx", sobelx)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(sobelx, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

sobely = cv.Sobel(img, cv.CV_16S, 0, 1, ksize=3)

cv.imshow("sobely", sobely)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(sobely, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

edges = cv.Canny(img, 10, 400)

cv.imshow("edges", edges)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(edges, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb, cmap="gray")
```
