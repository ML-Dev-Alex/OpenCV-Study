```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("pic.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img, cmap="gray", interpolation="bicubic")
```

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))
while True:
    ret, frame = cap.read()
    out.write(frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("frame", frame)
    cv2.imshow("gray", gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
color = cv2.imread("pic.jpg", cv2.IMREAD_COLOR)
cv2.line(color, (0, 1390), (2327, 1390), (255, 255, 255), 100)
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(color, "IN STERQUILINIIS INVENITUR", (240, 1430), font, 4, (0, 0, 0), 5, cv2.LINE_AA)
cv2.imshow("color", color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
