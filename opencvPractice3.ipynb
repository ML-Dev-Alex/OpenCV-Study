```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('page.jpg')
cv.imshow("original", img)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(img, dsize=(130, 240), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()


retval, threshold = cv.threshold(img, 85, 255, cv.THRESH_BINARY)

cv.imshow("threshold", threshold)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(threshold, dsize=(130, 240), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

grayscaled = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

retval2, threshold2 = cv.threshold(grayscaled, 85, 255, cv.THRESH_BINARY)

cv.imshow("threshold2", threshold2)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(threshold2, dsize=(130, 240), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb, cmap="gray")
```

```python
plt.clf()
plt.cla()
plt.close()

gaus = cv.adaptiveThreshold(grayscaled, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 285, 1)

cv.imshow("gaus", gaus)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(gaus, dsize=(130, 240), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb, cmap="gray")  # On the thumbnail it isn't apparent, but this is MUCH clearer to read at full size
```

```python
plt.clf()
plt.cla()
plt.close()

retval2, otsu = cv.threshold(grayscaled, 5, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)

cv.imshow("otsu", otsu)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(otsu, dsize=(130, 240), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb, cmap="gray")  
```
