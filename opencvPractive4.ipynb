```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('woman.jpg')

cv.imshow("original", img)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(img, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

lower_red = np.array([0, 0, 90])
upper_red = np.array([255, 255, 215])

mask = cv.inRange(hsv, lower_red, upper_red)
mask = cv.bitwise_not(mask)
res = cv.bitwise_and(img, img, mask=mask)


cv.imshow("mask", mask)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(mask, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb, cmap="gray")


```

```python
plt.clf()
plt.cla()
plt.close()


cv.imshow("res", res)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(res, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

kernel = np.ones((15, 15), np.float32) / 255
smoothed = cv.filter2D(res, -1, kernel)
cv.imshow("smoothed", smoothed)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(smoothed, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

blur = cv.GaussianBlur(res, (15, 15), 0)
cv.imshow("blur", blur)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(blur, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

median = cv.medianBlur(res, 15)
cv.imshow("median", median)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(median, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

kernel = np.ones((5, 5), np.uint8)
erosion = cv.erode(mask, kernel, iterations=1)
erosion = cv.bitwise_and(img, img, mask=erosion)
cv.imshow("erosion", erosion)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(erosion, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

kernel = np.ones((5, 5), np.uint8)
dilation = cv.dilate(mask, kernel, iterations=1)
dilation = cv.bitwise_and(img, img, mask=dilation)
cv.imshow("dilation", dilation)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(dilation, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])
```

```python
plt.clf()
plt.cla()
plt.close()

kernel = np.ones((5, 5), np.uint8)

opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)

opening = cv.bitwise_and(img, img, mask=opening)
cv.imshow("opening", opening)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(opening, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])

```

```python
plt.clf()
plt.cla()
plt.close()

kernel = np.ones((5, 5), np.uint8)

close = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)
close = cv.bitwise_and(img, img, mask=close)
cv.imshow("close", close)
cv.waitKey(0)
cv.destroyAllWindows()

thumb = cv.resize(close, dsize=(400, 300), interpolation=cv.INTER_CUBIC)
plt.imshow(thumb[:, :, ::-1])

```
